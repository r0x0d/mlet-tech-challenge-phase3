## Verificar quantidade de dados com popularidade 0 no dataset

```python
# Porcentagem de músicas com popularide 0
percentual_popularidade = len(df[df.popularity == 0]) / len(df) * 100
print("Percentual de músicas com popularide 0: {0:.2f}%".format(percentual_popularidade))

df.sort_values('popularity', ascending=False).reset_index().tail()
```

## Removendo as músicas com popularidade 0 para limpar o dataset

```python
# Atualiza dataset para remover músicas com popularidade 0, como visto na
# célula anterior, chegamos a 14% de músicas com baixa popularidade.
# Para não afetar o treinamento, optamos por remover tais músicas.
dataset_cleaned = df[df.popularity > 0]

print(dataset_cleaned.info())
dataset_cleaned.tail()
```

```python
# Verifica novamente a porcentagem de músicas com popularidade 0
percentual_popularidade = len(dataset_cleaned[dataset_cleaned.popularity == 0]) / len(dataset_cleaned) * 100
print("Percentual de músicas com popularide 0: {0:.2f}%".format(percentual_popularidade))
```

## Para o dataset limpo, essas colunas não são tão relevantes. Queremos trabalhar com um modelo numérico.

```python
dataset_cleaned.drop(['Unnamed: 0', 'track_id', 'artists', 'album_name', 'track_name', 'track_genre'], axis=1, inplace=True)
```

## Selecionando todas as colunas que contém valores numéricos como features

```python
features = df.select_dtypes(include=['number']).columns

features
```

Método **Z-Score** para verificar desvio dos dados em relação a média para identificar outliers.

```python
from scipy import stats
import numpy as np


# Calcular o Z-Score para cada valor das variáveis numéricas
z = np.abs(stats.zscore(dataset_cleaned[features]))

# Definir um limiar para considerar outliers (geralmente z > 3)
df_no_outliers = dataset_cleaned[(z < 3).all(axis=1)]

df_no_outliers.info()
```

Método **intervalo interquartil (IQR)**, que considera os valores que estão fora de 1.5 vezes o IQR como outliers.

O IQR é a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1), e valores que estão muito acima de Q3 ou muito abaixo de Q1 são considerados outliers.

```python
# Calcular os limites de IQR
Q1 = dataset_cleaned[features].quantile(0.25)
Q3 = dataset_cleaned[features].quantile(0.75)
IQR = Q3 - Q1

# Remover outliers (valores fora do intervalo de 1.5 vezes o IQR)
df_no_outliers = dataset_cleaned[~((dataset_cleaned[features] < (Q1 - 1.5 * IQR)) | (dataset_cleaned[features] > (Q3 + 1.5 * IQR))).any(axis=1)]

df_no_outliers.info()
```

Visualizar os outliers usando gráficos de boxplot para verificar quais variáveis podem ter outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criar boxplots para visualizar os outliers em cada variável
plt.figure(figsize=(10, 8))
sns.boxplot(data=df_no_outliers[features])
plt.xticks(rotation=90)
plt.title("Boxplot para identificação de outliers")
plt.show()
```
