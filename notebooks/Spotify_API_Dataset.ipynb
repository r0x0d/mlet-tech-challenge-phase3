### Carregar o arquivo com o Pandas e exibir o resumo.

```python
import pandas as pd

# Tentar carregar o arquivo especificando a codificação
df = pd.read_csv('spotify_data.csv', encoding='latin1')

# Exibir as primeiras linhas do dataset para verificar o carregamento
df.head()
```

```python
# Informações sobre o dataset (colunas, tipos de dados, valores nulos)
df.info()
```

```python
# Descrição estatística dos dados numéricos
df.describe()
```

```python
# Verificar valores nulos
df.isnull().sum()
```

Converte o tempo de duração de uma música de milisegundos para minutos.

```python
# Cria uma nova coluna chamada "duration_mins" para armazenar a duração das músicas em minutos ao invés de milisegundos.
df['duration_mins'] = df['duration_ms'] / 60000

df['duration_mins']

df = df.drop(['track_id', 'duration_ms'], axis=1)
```

```python
# Verificando que não temos nenhum dado duplicado no dataset.
df.duplicated().sum() == 0

df.duplicated()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cmap = sns.diverging_palette(230, 20, as_cmap=True)

plt.figure(figsize=(15, 10))

attributes = df[['popularity', 'year', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'duration_mins']]
sns.heatmap(attributes.corr(), annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap=cmap)
plt.title("Matriz de correlações", fontweight='bold', fontsize='large')
```

Criando a coluna de popularidade de acordo com a quantidade de streams.

```python
# Criar categorias com base em quantis (33% de cada faixa)

df = df.copy()
df['popularity_category'] = ''

for index, row in df.iterrows():
    score = 'Baixo'
    if row.popularity >= 50:
      score = 'Alto'

    df.at[index, 'popularity_category'] = score

# Exibir as primeiras linhas para verificar o resultado
df[['artist_name', 'track_name', 'popularity', 'popularity_category']].head()
```

```python
# Porcentagem de músicas com popularide 0
percentual_popularidade = len(df[df.popularity == 0]) / len(df) * 100
print("Percentual de músicas com popularide 0: {0:.2f}%".format(percentual_popularidade))

df.sort_values('popularity', ascending=False).reset_index().tail()
```

```python
# Atualiza dataset para remover músicas com popularidade 0, como visto na
# célula anterior, chegamos a quase 14% de músicas com baixa popularidade.
# Para não afetar o treinamento, optamos por remover tais músicas.
dataset_cleaned = df[df.popularity > 0]

print(dataset_cleaned.info())
dataset_cleaned.tail()
```

```python
features = ['year', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'duration_mins']
```

Método **Z-Score** para verificar desvio dos dados em relação a média para identificar outliers.

```python
from scipy import stats
import numpy as np


# Calcular o Z-Score para cada valor das variáveis numéricas
z = np.abs(stats.zscore(dataset_cleaned[features]))

# Definir um limiar para considerar outliers (geralmente z > 3)
df_no_outliers = dataset_cleaned[(z < 3).all(axis=1)]

df_no_outliers.info()
```

Método **intervalo interquartil (IQR)**, que considera os valores que estão fora de 1.5 vezes o IQR como outliers.

O IQR é a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1), e valores que estão muito acima de Q3 ou muito abaixo de Q1 são considerados outliers.

```python
# Calcular os limites de IQR
Q1 = dataset_cleaned[features].quantile(0.25)
Q3 = dataset_cleaned[features].quantile(0.75)
IQR = Q3 - Q1

# Remover outliers (valores fora do intervalo de 1.5 vezes o IQR)
df_no_outliers = dataset_cleaned[~((dataset_cleaned[features] < (Q1 - 1.5 * IQR)) | (dataset_cleaned[features] > (Q3 + 1.5 * IQR))).any(axis=1)]

df_no_outliers.info()
```

Visualizar os outliers usando gráficos de boxplot para verificar quais variáveis podem ter outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criar boxplots para visualizar os outliers em cada variável
plt.figure(figsize=(10, 8))
sns.boxplot(data=df_no_outliers[features])
plt.xticks(rotation=90)
plt.title("Boxplot para identificação de outliers")
plt.show()
```

```python
# Distribuição por populariade no dataset

df_no_outliers['popularity'].plot(kind='hist', bins=50)
plt.xlabel('Score')
plt.ylabel('Número de músicas')
plt.title('Distribuição por popularidade')
```

```python
# Popularidade vs Volume
fig, ax = plt.subplots(figsize= (15, 10))
sns.scatterplot(data=df_no_outliers, x = 'popularity', y = 'loudness', hue = 'popularity_category', size = 'popularity_category', sizes = (20, 200), legend = 'full')
```

```python
#Popularidade vs Fato dançante
fig, ax = plt.subplots(figsize= (15, 10))
sns.scatterplot(data = df_no_outliers, x = 'popularity', y = 'danceability', hue = 'popularity_category', size = 'popularity_category', sizes = (20, 200), legend = 'full')
```
